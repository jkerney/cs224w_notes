### Visualise data from ogbg-molhiv dataset

```python
import torch_geometric
import networkx as nx
import matplotlib.pyplot as plt
from ogb.graphproppred import PygGraphPropPredDataset
```

```python
# Load the dataset
dataset = PygGraphPropPredDataset(name='ogbg-molhiv')
```

```python
# Select a subset of graphs to visualize (e.g., the first 10 graphs)
graphs = dataset[:20]

#Loop over the selected graphs and visualize them using NetworkX
for i, graph in enumerate(graphs):
    # Extract the graph label
    label = graph.y.item()

    # Convert the graph to a NetworkX graph
    nx_graph = torch_geometric.utils.to_networkx(graph)

    # Draw the NetworkX graph using a circular layout
    pos = nx.circular_layout(nx_graph)
    nx.draw(nx_graph, pos, with_labels=True)
    plt.title(f"ogbg-molhiv: Graph {i+1}, Label: {label}")
    plt.show()
```
